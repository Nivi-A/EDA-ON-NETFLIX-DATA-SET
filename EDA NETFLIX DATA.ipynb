# PRE PROCESSING : data cleaning

```python
#number of unique elements
nf.nunique()
```

```python
#Detect missing values
nf.isna().sum()
```

```python
#PERCENTAGE OF NULL DATAS:
missing_values_count = nf.isnull().sum()
# how many total missing values do we have?
total_cells = np.product(nf.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)

#4 percent is missing
```

```python
#no duplicate  datas to delete
nf.duplicated().sum()
```

```python
# remove all the rows that contain a missing value
nf.dropna()
#WHICH MEANS ALL ROWS HAVE ATLEAST ONE NULL VALUE
```

**The missing values are in:**

"director": drop the rows with null values

"cast": there are too many diferent values so drop this

"country":Important variable , so we will fix this with mode or some other values

"date_added": there are just a few cases, so lets assign 0

"rating": there are just a few cases, so lets replace it with mode

```python
# APPROACHES TO DEAL WITH MISSING DATA 

'''
1. Drop column : we can drop this col if we want since it has large number of missing values(more than quater)
    nf.drop('director', axis=1)
2. droping empty row on the basis of the given column
    (if axis = 0 - > drops the rows ; if axis = 1 -> drops the columns)
    nf.dropna(subset=['director'])'''
```

```python
# 3 dropping row : we can drop all the rows corresponding this column (applied this)
director_no_null = nf[nf['director'].isnull()].index
nf.drop(director_no_null, axis=0 , inplace=True)
nf
```

```python
#we drop cast rows where it is null
cast_no_nul = nf[nf['cast'].isnull()].index
nf.drop(cast_no_nul, axis=0 , inplace=True)
nf
```

```python
#country is filled with mode 
nf['country'] = nf['country'].fillna(nf['country'].mode()[0])
```

```python
# Lets retrieve just the first country
nf['main_country'] = nf['country'].apply(lambda x: x.split(",")[0])
nf['main_country'].head()
```

```python
# rating has only 4 null values and also its categorical , so it can be replaced with the mode of that entire column
mode = ''.join(nf['rating'].mode())
nf['rating'].fillna(mode, inplace=True)
```

```python
#so null values of rating column will be filled with 'TV-MA'
mode
```

```python
# we have only 3 null values in duration , so we can better replace those with 0 
#and also its data type is string , so we cant take mean or median of those duration so 0 would be better
#we can also use ffill or bfill but i think for duration column 0 suits the best 
nf['duration'].fillna(0, inplace=True)
```

```python
#FINALLY WE ONLY HAVE NULL VALUES IN 2 COLUMN THAT WE WILL TAKE CARE LATER
nf.isnull().sum()
```

```python
nf.shape
```

```python
#PERCENTAGE OF NULL DATAS:
missing_values_count = nf.isnull().sum()
# how many total missing values do we have?
total_cells = np.product(nf.shape)
total_missing = missing_values_count.sum()

# percent of data that is missing
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)

#after cleaning only 1 percentage of data is missing 
```

# WORKING WITH OUTLIERS

**outliers in attribute "duration"** :

```python
#working with outliers 
''' in the column duration we have values like 10 mins , 20 mins (n mins) for type movies and also like 
season 1 , season 2(season n) for type TV SHOWS
so to find the outliers in this column(duration) we only need the values where type = movies,so that we have duration in minutes '''
df_movie = nf[nf['type']=='Movie']
```

```python
#since duration is in object data type , we extract the numbers and convert into int datatype to find outliers 
df_movie['duration'] = df_movie['duration'].astype('str').str.extractall('(\d+)').unstack().fillna('').sum(axis=1).astype(int)
```

```python
#so this is how the duration column appears
df_movie['duration']
```

```python
# we 1st use barplot using matplotlib for fiinding outliers
fig, ax = plt.subplots(nrows=1, ncols=1)
plt.hist(df_movie['duration'])
fig.tight_layout()
'''it appears like the movies in the first bar and the last visible bar are probably outliers. 
They might be movies with very short time (short films) or very long documentaries which are put in movies category
and it dooesnt fit this category well''' 
```

```python
#its best to use boxplot to identify the outliers 
fig, ax = plt.subplots(nrows=1, ncols=1)
ax = sns.boxplot(x=df_movie['duration'])
fig.tight_layout()

#we can clearly see the number of outliers : nearly 50 below and 150 above are outliers
```

```python
#this helps us to know IQR , mean , median , mode etc
df_movie['duration'].describe()
```

```python
# now we will filter out the outliers 
#outliers (i took approximate range )
df_movie[(df_movie['duration']<40) | (df_movie['duration']>155)]
#filtering outliers out
df_movie = df_movie[(df_movie['duration']>40) & (df_movie['duration']<155)]
#The df_movie created now contains only movies that last between 40 and 155.
```

```python
# we are going to do barplot on rating so we will 1st see the distinct values of col rating
df_movie.rating.unique()
```

```python
#-----------------------------------------rating vs content type-heat map-----------------------------------------------------

colormap = plt.cm.plasma
sns.heatmap(pd.crosstab(nf["rating"], nf["type"]), cmap = colormap)
#we see mode of rating is TV-MA , as its repeated lot
```

```python
#clearly 74min , 84min , 66 min doesnt come under ratings and we can remove it or replace it with mode
nf['rating'] = nf['rating'].replace(['74 min','84 min','66 min'],'TV-MA')
```
